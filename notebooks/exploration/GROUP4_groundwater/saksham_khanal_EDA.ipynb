```python
for column in data.columns:
    plt.figure(figsize=(10,10))
    sns.histplot(data[column],kde=True)
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
data_null = data.isnull().sum()
data_null
```

```python
corr = data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Plot')
plt.show()
```

```python
data.date
```

```python
data['date'] = pd.to_datetime(data['date'],infer_datetime_format=True)
columns_to_plot = ['trees','water','grass','flooded_vegetation','crops','shrub_and_scrub','built','bare','snow_and_ice','label']
```

```python
plt.figure(figsize=(10,10))
for column in columns_to_plot:
    plt.scatter(data['date'],data[column],marker='o',label=column)
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.show()
```

```python
data.tail()
```

```python
grouped = data.groupby('city')
```

```python
grouped
```

```python

fig, axes = plt.subplots(nrows=len(grouped), ncols=1, figsize=(10, 6 * len(grouped)))

for i, (city, data) in enumerate(grouped):
    axes[i].plot(data['date'], data['trees'], label='Trees')
    axes[i].plot(data['date'], data['water'], label='Water')

    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Count')
    axes[i].set_title(city + ' - Trees and Water Time Series')
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
```
